## 筛选链接

之所以考虑获取 anki 中所有的图片链接，是想要将所有的链接图片下载下来，以防止由于官网更换图片渠道后，无法正常访问。

```python
import re
def main():
    obFile = open('公共基础文本.txt','r', encoding='UTF-8')
    text = obFile.read()
    obFile.close()
    obFile = open('公共基础文本urls.txt','w')
    pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    obFile.write('\n'.join(re.findall(pattern,text)))
    obFile.close()
if __name__ == '__main__':
    main()
```

## 下载图片

```python
import urllib.request
import re
```

```python
obFile = open('专业基础urls.txt','r', encoding='UTF-8')
imgname = ''
for i in obFile:
    imgurl = i.replace('\n','')
    imgname = imgurl.split('/')[-1]
    print(imgurl)
    urllib.request.urlretrieve(imgurl, '专业基础urls图片\\%s' % imgname)
obFile.close()    
```
